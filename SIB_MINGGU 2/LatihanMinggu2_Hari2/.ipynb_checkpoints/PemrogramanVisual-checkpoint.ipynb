```python
#List Box
def latihan_8():
    root = tk.Tk()
    root.title("Latihan 8: Listbox")
    listbox = tk.Listbox(root)
    listbox.insert(1, "Pilihan 1")
    listbox.insert(2, "Pilihan 2")
    listbox.insert(3, "Pilihan 3")
    listbox.pack()
    def pilih():
        pilihan = listbox.get(listbox.curselection())
        label_pilihan.config(text=f"Kamu memilih: (pilihan)")
    tombol = tk.Button(root, text="Pilih", command=pilih)
    tombol.pack()
    label_pilihan = tk. Label(root, text="")
    label_pilihan.pack()
    root.mainloop()
latihan_8()
```

```python
#Program Slider
def latihan_9():
    root = tk.Tk() 
    root.title("Latihan 9: Skala")
    
    def update_value(value): 
        label_value.config(text=f"Nilai: {value}")
        
    scale = tk.Scale (root, from_=0, to=100, orient=tk. HORIZONTAL, command=update_value) 
    scale.pack()
    
    label_value = tk. Label (root, text="Nilai: 0")
    label_value.pack()
    root.mainloop()
    
latihan_9()
```

```python
#Program Tabel data
from tkinter import ttk
def latihan_10():
    root = tk.Tk()
    root.title("Latihan 10: Tabel Data")
    tree = ttk. Treeview(root, columns=("Nama", "Umur", "Kota"), show='headings')
    tree.heading("Nama", text="Nama")
    tree.heading("Umur", text="Umur")
    tree.heading("Kota", text="Kota")
    data = [ 
        ("Budi", 25, "Jakarta"), 
        ("Susi", 30, "Bandung"), 
        ("Andi", 22, "Surabaya"),
    ]
    for row in data:
        tree.insert("", tk.END, values=row)
    tree.pack()
    root.mainloop()
    
latihan_10()
```

```python
#Program To-Do-List
def latihan_11():
    root = tk.Tk()
    root.title("Latihan 11: To-Do List")
    tasks = []
    def tambah_tugas():
        tugas = entry_tugas.get()
        if tugas:
            tasks.append(tugas)
            update_listbox()
            entry_tugas.delete(0, tk.END)
    def hapus_tugas():
        selected_task = listbox_tugas.curselection()
        if selected_task:
            tasks.pop(selected_task[0])
            update_listbox()
    def update_listbox():
        listbox_tugas.delete(0, tk.END)
        for task in tasks:
            listbox_tugas.insert(tk.END, task)
    entry_tugas = tk. Entry (root)
    entry_tugas.pack() 
    tombol_tambah = tk.Button(root, text="Tambah Tugas", command=tambah_tugas)
    tombol_tambah.pack()
    tombol_hapus = tk.Button (root, text="Hapus Tugas", command=hapus_tugas)
    tombol_hapus.pack()
    listbox_tugas = tk.Listbox(root)
    listbox_tugas.pack()
    root.mainloop()
latihan_11()
```

```python
# LATIHAN 1 PEMROGRAMAN VISUSAL
# 1. Menghitung volume balok
def latihan_1():
    root = tk.Tk()
    root.title("Latihan 1: Kalkulator Volume Balok")
    def hitung_volume():
        panjang = float(entry_panjang.get())
        lebar = float(entry_lebar.get())
        tinggi = float(entry_tinggi.get())
        volume = panjang * lebar * tinggi
        label_hasil.config(text=f"Volume: {volume} m³")
    # Input untuk Panjang
    tk.Label(root, text="Panjang (m):").pack()
    entry_panjang = tk.Entry(root)
    entry_panjang.pack()
    # Input untuk Lebar
    tk.Label(root, text="Lebar (m):").pack()
    entry_lebar = tk.Entry(root)
    entry_lebar.pack()
    # Input untuk Tinggi
    tk.Label(root, text="Tinggi (m):").pack()
    entry_tinggi = tk.Entry(root)
    entry_tinggi.pack()
    # Tombol Hitung
    tombol_hitung = tk.Button(root, text="Hitung Volume", command=hitung_volume)
    tombol_hitung.pack()
    # Label Hasil
    label_hasil = tk.Label(root, text="Volume: ")
    label_hasil.pack()
    root.mainloop()

latihan_1()
```

```python
# 2. Menentukan Bilangan Prima 
def latihan_2():
    root = tk.Tk()
    root.title("Latihan 2: Cek Bilangan Prima")
    def cek_prima():
        try:
            angka = int(entry_angka.get())
            if angka > 1:
                for i in range(2, angka):
                    if (angka % i) == 0:
                        label_hasil.config(text=f"{angka} bukan bilangan prima.")
                        break
                else:
                    label_hasil.config(text=f"{angka} adalah bilangan prima.")
            else:
                label_hasil.config(text=f"{angka} bukan bilangan prima.")
        except ValueError:
            label_hasil.config(text="Masukkan angka yang valid.")
    # Input Angka
    tk.Label(root, text="Masukkan Angka:").pack()
    entry_angka = tk.Entry(root)
    entry_angka.pack()
    # Tombol Cek Prima
    tombol_cek = tk.Button(root, text="Cek Prima", command=cek_prima)
    tombol_cek.pack()
    # Label Hasil
    label_hasil = tk.Label(root, text="")
    label_hasil.pack()
    root.mainloop()

latihan_2()
```
