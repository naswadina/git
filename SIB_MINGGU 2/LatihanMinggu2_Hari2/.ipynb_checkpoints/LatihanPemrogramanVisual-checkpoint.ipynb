```python
# Aplikasi Kalkulator Sederhana
def kalkulator():
    root = tk.Tk()
    root.title("Kalkulator Sederhana")
    def hitung(operasi):
        try:
            angka1 = float(entry_angka1.get())
            angka2 = float(entry_angka2.get())
            if operasi == '+':
                hasil = angka1 + angka2
            elif operasi == '-':
                hasil = angka1 - angka2
            elif operasi == '*':
                hasil = angka1 * angka2
            elif operasi == '/':
                if angka2 != 0:
                    hasil = angka1 / angka2
                else:
                    hasil = "Tidak dapat dibagi dengan 0"
            label_hasil.config(text=f"Hasil: {hasil}")
        except ValueError:
            label_hasil.config(text="Masukkan angka yang valid")
    # Label dan Entry untuk Angka 1
    tk.Label(root, text="Angka 1:").pack()
    entry_angka1 = tk.Entry(root)
    entry_angka1.pack()
    # Label dan Entry untuk Angka 2
    tk.Label(root, text="Angka 2:").pack()
    entry_angka2 = tk.Entry(root)
    entry_angka2.pack()
    # Tombol Penjumlahan
    tk.Button(root, text="Tambah (+)", command=lambda: hitung('+')).pack()
    # Tombol Pengurangan
    tk.Button(root, text="Kurang (-)", command=lambda: hitung('-')).pack()
    # Tombol Perkalian
    tk.Button(root, text="Kali (*)", command=lambda: hitung('*')).pack()
    # Tombol Pembagian
    tk.Button(root, text="Bagi (/)", command=lambda: hitung('/')).pack()
    # Label untuk menampilkan hasil
    label_hasil = tk.Label(root, text="Hasil: ")
    label_hasil.pack()
    root.mainloop()

kalkulator()
```

```python
# To-Do List dengan Opsi Simpan dan Buka
from tkinter import filedialog
import os

def list():
    root = tk.Tk()
    root.title("To-Do List")

    tasks = []

    # Fungsi untuk menambahkan tugas ke daftar
    def tambah_tugas():
        tugas = entry_tugas.get()
        if tugas:
            tasks.append(tugas)
            update_listbox()
            entry_tugas.delete(0, tk.END)

    # Fungsi untuk menghapus tugas yang dipilih
    def hapus_tugas():
        selected_task = listbox_tugas.curselection()
        if selected_task:
            tasks.pop(selected_task[0])
            update_listbox()

    # Fungsi untuk menyimpan daftar tugas ke file teks
    def simpan_tugas():
        file_path = filedialog.asksaveasfilename(
            defaultextension=".txt", filetypes=[("Text Files", "*.txt")]
        )
        if file_path:
            with open(file_path, 'w') as file:
                for task in tasks:
                    file.write(task + '\n')
            print(f"Tugas disimpan di {file_path}")

    # Fungsi untuk membuka daftar tugas dari file teks
    def buka_tugas():
        file_path = filedialog.askopenfilename(
            filetypes=[("Text Files", "*.txt")]
        )
        if file_path:
            tasks.clear()
            with open(file_path, 'r') as file:
                for line in file:
                    tasks.append(line.strip())
            update_listbox()
            print(f"Tugas dibuka dari {file_path}")

    # Fungsi untuk memperbarui Listbox sesuai dengan daftar tugas
    def update_listbox():
        listbox_tugas.delete(0, tk.END)
        for task in tasks:
            listbox_tugas.insert(tk.END, task)

    # Entry untuk memasukkan tugas
    entry_tugas = tk.Entry(root)
    entry_tugas.pack()

    # Tombol untuk menambahkan tugas
    tombol_tambah = tk.Button(root, text="Tambah Tugas", command=tambah_tugas)
    tombol_tambah.pack()

    # Tombol untuk menghapus tugas
    tombol_hapus = tk.Button(root, text="Hapus Tugas", command=hapus_tugas)
    tombol_hapus.pack()

    # Listbox untuk menampilkan daftar tugas
    listbox_tugas = tk.Listbox(root)
    listbox_tugas.pack()

    # Tombol untuk menyimpan daftar tugas
    tombol_simpan = tk.Button(root, text="Simpan Tugas", command=simpan_tugas)
    tombol_simpan.pack()

    # Tombol untuk membuka daftar tugas dari file
    tombol_buka = tk.Button(root, text="Buka Tugas", command=buka_tugas)
    tombol_buka.pack()

    root.mainloop()

list()
```

```python
def buku_telepon():
    root = tk.Tk()
    root.title("Buku Telepon")
    contacts = {}
    # Fungsi untuk menambah kontak
    def tambah_kontak():
        nama = entry_nama.get()
        nomor = entry_nomor.get()
        if not nama:
            messagebox.showerror("Error", "Nama tidak boleh kosong.")
            return
        if not nomor.isdigit():
            messagebox.showerror("Error", "Nomor telepon hanya berisi angka.")
            return
        contacts[nama] = nomor
        update_listbox()
        entry_nama.delete(0, tk.END)
        entry_nomor.delete(0, tk.END)
    # Fungsi untuk mengedit kontak
    def edit_kontak():
        selected_contact = listbox_kontak.curselection()
        if selected_contact:
            nama = listbox_kontak.get(selected_contact).split(":")[0].strip()  # Mengambil nama dari listbox
            if nama in contacts:
                nomor = contacts[nama]
                entry_nama.delete(0, tk.END)
                entry_nomor.delete(0, tk.END)
                entry_nama.insert(0, nama)
                entry_nomor.insert(0, nomor)
                # Menghapus kontak dari dictionary untuk memperbarui
                del contacts[nama]
                update_listbox()
    # Fungsi untuk menghapus kontak
    def hapus_kontak():
        selected_contact = listbox_kontak.curselection()
        if selected_contact:
            nama = listbox_kontak.get(selected_contact).split(":")[0].strip()  # Mengambil nama dari listbox
            if nama in contacts:
                del contacts[nama]
                update_listbox()
    # Fungsi untuk memperbarui Listbox
    def update_listbox():
        listbox_kontak.delete(0, tk.END)
        for nama, nomor in contacts.items():
            listbox_kontak.insert(tk.END, f"{nama}: {nomor}")
    # Label dan Entry untuk Nama
    tk.Label(root, text="Nama:").pack()
    entry_nama = tk.Entry(root)
    entry_nama.pack()
    # Label dan Entry untuk Nomor Telepon
    tk.Label(root, text="Nomor Telepon:").pack()
    entry_nomor = tk.Entry(root)
    entry_nomor.pack()
    # Tombol untuk Menambah Kontak
    tombol_tambah = tk.Button(root, text="Tambah Kontak", command=tambah_kontak)
    tombol_tambah.pack()
    # Tombol untuk Mengedit Kontak
    tombol_edit = tk.Button(root, text="Edit Kontak", command=edit_kontak)
    tombol_edit.pack()
    # Tombol untuk Menghapus Kontak
    tombol_hapus = tk.Button(root, text="Hapus Kontak", command=hapus_kontak)
    tombol_hapus.pack()
    # Listbox untuk Menampilkan Kontak
    listbox_kontak = tk.Listbox(root)
    listbox_kontak.pack()
    root.mainloop()

buku_telepon()
```

```python
# Aplikasi Voting dengan Grafik Batang
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
def voting():
    root = tk.Tk()
    root.title("Aplikasi Voting Sederhana")
    votes = {"Kandidat 1": 0, "Kandidat 2": 0, "Kandidat 3": 0}  # Inisialisasi jumlah suara untuk setiap kandidat
    def update_graph():  # Fungsi untuk memperbarui grafik batang
        fig, ax = plt.subplots()
        ax.bar(votes.keys(), votes.values(), color=['blue', 'green', 'red'])
        ax.set_xlabel("Kandidat")
        ax.set_ylabel("Jumlah Suara")
        ax.set_title("Hasil Voting")
        for widget in frame_graph.winfo_children(): # Hapus kanvas lama dan buat yang baru
            widget.destroy()
        canvas = FigureCanvasTkAgg(fig, master=frame_graph)
        canvas.draw()
        canvas.get_tk_widget().pack()
    def vote(candidate):  # Fungsi untuk memberikan suara kepada kandidat
        votes[candidate] += 1
        messagebox.showinfo("Terima Kasih!", f"Anda memilih {candidate}")
        update_graph()
    tk.Label(root, text="Pilih Kandidat:").pack()  # Label instruksi
    tk.Button(root, text="Kandidat 1", command=lambda: vote("Kandidat 1")).pack()  # Tombol untuk memberikan suara kepada kandidat
    tk.Button(root, text="Kandidat 2", command=lambda: vote("Kandidat 2")).pack()
    tk.Button(root, text="Kandidat 3", command=lambda: vote("Kandidat 3")).pack()
    frame_graph = tk.Frame(root)  # Frame untuk menampilkan grafik
    frame_graph.pack()
    update_graph()  # Tampilkan grafik awal
    root.mainloop()

voting()
```

```python
# Buku Penyewaan Film dengan Penyaringan
from tkinter import messagebox

# Aplikasi Penyewaan Film
def film():
    root = tk.Tk()
    root.title("Aplikasi Penyewaan Film")

    # Data Film (judul, genre, status)
    film_list = [
        {"judul": "Film 1", "genre": "Action", "status": "Tersedia"},
        {"judul": "Film 2", "genre": "Drama", "status": "Tersedia"},
        {"judul": "Film 3", "genre": "Comedy", "status": "Tersedia"},
    ]

    genres = ["Semua", "Action", "Drama", "Comedy"]

    # Fungsi untuk memperbarui daftar film berdasarkan filter genre
    def update_film_list():
        selected_genre = genre_var.get()
        listbox_film.delete(0, tk.END)
        for film in film_list:
            if selected_genre == "Semua" or film["genre"] == selected_genre:
                listbox_film.insert(tk.END, f"{film['judul']} ({film['genre']}) - {film['status']}")

    # Fungsi untuk menyewa film
    def sewa_film():
        selected_film_index = listbox_film.curselection()
        if selected_film_index:
            selected_film_text = listbox_film.get(selected_film_index)
            film_judul = selected_film_text.split(" (")[0]
            for film in film_list:
                if film["judul"] == film_judul:
                    if film["status"] == "Tersedia":
                        film["status"] = "Disewa"
                        messagebox.showinfo("Sukses", f"Anda menyewa {film_judul}")
                    else:
                        messagebox.showerror("Error", "Film sudah disewa")
                    update_film_list()
                    return

    # Fungsi untuk mengembalikan film
    def kembalikan_film():
        selected_film_index = listbox_film.curselection()
        if selected_film_index:
            selected_film_text = listbox_film.get(selected_film_index)
            film_judul = selected_film_text.split(" (")[0]
            for film in film_list:
                if film["judul"] == film_judul:
                    if film["status"] == "Disewa":
                        film["status"] = "Tersedia"
                        messagebox.showinfo("Sukses", f"Anda mengembalikan {film_judul}")
                    else:
                        messagebox.showerror("Error", "Film belum disewa")
                    update_film_list()
                    return

    # Fungsi untuk menambah film baru
    def tambah_film():
        judul = entry_judul.get()
        genre = genre_var_tambah.get()
        if judul and genre != "Pilih Genre":
            film_list.append({"judul": judul, "genre": genre, "status": "Tersedia"})
            update_film_list()
            entry_judul.delete(0, tk.END)
        else:
            messagebox.showerror("Error", "Judul atau genre tidak boleh kosong")

    # Fungsi untuk menghapus film dari daftar
    def hapus_film():
        selected_film_index = listbox_film.curselection()
        if selected_film_index:
            selected_film_text = listbox_film.get(selected_film_index)
            film_judul = selected_film_text.split(" (")[0]
            for i, film in enumerate(film_list):
                if film["judul"] == film_judul:
                    del film_list[i]
                    update_film_list()
                    messagebox.showinfo("Sukses", f"{film_judul} telah dihapus")
                    return

    # Frame Filter dan Daftar Film
    frame_film = tk.Frame(root)
    frame_film.pack(pady=10)

    # Dropdown untuk memilih genre
    genre_var = tk.StringVar(value="Semua")
    genre_dropdown = tk.OptionMenu(frame_film, genre_var, *genres, command=lambda x: update_film_list())
    genre_dropdown.pack()

    # Listbox untuk menampilkan daftar film
    listbox_film = tk.Listbox(frame_film, width=50)
    listbox_film.pack()

    # Tombol Sewa dan Kembalikan
    btn_sewa = tk.Button(frame_film, text="Sewa", command=sewa_film)
    btn_sewa.pack(side=tk.LEFT, padx=5)
    btn_kembalikan = tk.Button(frame_film, text="Kembalikan", command=kembalikan_film)
    btn_kembalikan.pack(side=tk.LEFT, padx=5)

    # Frame Tambah dan Hapus Film
    frame_manage = tk.Frame(root)
    frame_manage.pack(pady=10)

    tk.Label(frame_manage, text="Judul Film:").pack()
    entry_judul = tk.Entry(frame_manage)
    entry_judul.pack()

    genre_var_tambah = tk.StringVar(value="Pilih Genre")
    genre_dropdown_tambah = tk.OptionMenu(frame_manage, genre_var_tambah, *genres[1:])
    genre_dropdown_tambah.pack()

    btn_tambah = tk.Button(frame_manage, text="Tambah Film", command=tambah_film)
    btn_tambah.pack(side=tk.LEFT, padx=5)
    btn_hapus = tk.Button(frame_manage, text="Hapus Film", command=hapus_film)
    btn_hapus.pack(side=tk.LEFT, padx=5)

    # Update listbox dengan film awal
    update_film_list()

    root.mainloop()

film()
```

```python
# Kalkulator BMI (Body Mass Index)
from tkinter import messagebox

def kalkulator_bmi():
    root = tk.Tk()
    root.title("Kalkulator BMI")

    # Fungsi untuk menghitung BMI dan menampilkan hasil
    def hitung_bmi():
        try:
            berat = float(entry_berat.get())
            tinggi = float(entry_tinggi.get())
            if tinggi <= 0:
                messagebox.showerror("Error", "Tinggi badan harus lebih dari 0")
                return
            bmi = berat / (tinggi ** 2)
            kategori = ""
            if bmi < 18.5:
                kategori = "Underweight"
            elif 18.5 <= bmi < 24.9:
                kategori = "Normal weight"
            elif 25 <= bmi < 29.9:
                kategori = "Overweight"
            else:
                kategori = "Obese"
            label_hasil.config(text=f"BMI: {bmi:.2f}\nKategori: {kategori}")
        except ValueError:
            messagebox.showerror("Error", "Harap masukkan angka yang valid")

    # Label dan Entry untuk Berat Badan
    tk.Label(root, text="Berat Badan (kg):").pack()
    entry_berat = tk.Entry(root)
    entry_berat.pack()

    # Label dan Entry untuk Tinggi Badan
    tk.Label(root, text="Tinggi Badan (m):").pack()
    entry_tinggi = tk.Entry(root)
    entry_tinggi.pack()

    # Tombol untuk menghitung BMI
    tombol_hitung = tk.Button(root, text="Hitung BMI", command=hitung_bmi)
    tombol_hitung.pack()

    # Label untuk menampilkan hasil BMI dan kategori
    label_hasil = tk.Label(root, text="BMI: \nKategori: ")
    label_hasil.pack()

    root.mainloop()

kalkulator_bmi()
```

```python
# Aplikasi Konversi Mata Uang
def konversi_mata_uang():
    root = tk.Tk()
    root.title("Konversi Mata Uang")
    # Nilai tukar tetap (contoh: 1 USD = 15.000 IDR, 1 EUR = 16.000 IDR)
    tukar_mata_uang = {
        ('USD', 'IDR'): 15000,
        ('IDR', 'USD'): 1 / 15000,
        ('EUR', 'IDR'): 16000,
        ('IDR', 'EUR'): 1 / 16000,
        ('USD', 'EUR'): 0.9,  # Contoh nilai tukar USD ke EUR
        ('EUR', 'USD'): 1 / 0.9,
    }
    # Fungsi untuk melakukan konversi
    def konversi():
        try:
            jumlah = float(entry_jumlah.get())
            mata_uang_asal = mata_uang_asal_var.get()
            mata_uang_tujuan = mata_uang_tujuan_var.get()
            if mata_uang_asal == mata_uang_tujuan:
                hasil = jumlah
            else:
                tukar = tukar_mata_uang.get((mata_uang_asal, mata_uang_tujuan))
                if tukar is not None:
                    hasil = jumlah * tukar
                else:
                    messagebox.showerror("Error", "Nilai tukar tidak tersedia")
                    return
            label_hasil.config(text=f"Hasil Konversi: {hasil:.2f} {mata_uang_tujuan}")
        except ValueError:
            messagebox.showerror("Error", "Harap masukkan jumlah uang yang valid")
    # Label dan Entry untuk jumlah uang
    tk.Label(root, text="Jumlah Uang:").pack()
    entry_jumlah = tk.Entry(root)
    entry_jumlah.pack()
    # Dropdown untuk memilih mata uang asal
    tk.Label(root, text="Mata Uang Asal:").pack()
    mata_uang_asal_var = tk.StringVar(value="USD")
    mata_uang_asal_dropdown = tk.OptionMenu(root, mata_uang_asal_var, "USD", "IDR", "EUR")
    mata_uang_asal_dropdown.pack()
    # Dropdown untuk memilih mata uang tujuan
    tk.Label(root, text="Mata Uang Tujuan:").pack()
    mata_uang_tujuan_var = tk.StringVar(value="IDR")
    mata_uang_tujuan_dropdown = tk.OptionMenu(root, mata_uang_tujuan_var, "USD", "IDR", "EUR")
    mata_uang_tujuan_dropdown.pack()
    # Tombol untuk melakukan konversi
    tombol_konversi = tk.Button(root, text="Konversi", command=konversi)
    tombol_konversi.pack()
    # Label untuk menampilkan hasil konversi
    label_hasil = tk.Label(root, text="Hasil Konversi: ")
    label_hasil.pack()
    root.mainloop()

konversi_mata_uang()
```

```python
import random

def pembelajaran_matematika():
    root = tk.Tk()
    root.title("Pembelajaran Matematika")

    # Variabel global untuk menyimpan angka dan jawaban yang benar
    num1 = num2 = correct_answer = 0

    # Fungsi untuk menghasilkan angka acak dan jenis operasi
    def generate_problem(*args):
        nonlocal num1, num2, correct_answer
        num1 = random.randint(1, 20)
        num2 = random.randint(1, 20)
        if operasi_var.get() == "Penjumlahan":
            correct_answer = num1 + num2
            question.set(f"{num1} + {num2} = ?")
        elif operasi_var.get() == "Pengurangan":
            correct_answer = num1 - num2
            question.set(f"{num1} - {num2} = ?")

    # Fungsi untuk memeriksa jawaban pengguna
    def cek_jawaban():
        try:
            user_answer = int(entry_jawaban.get())
            if user_answer == correct_answer:
                messagebox.showinfo("Hasil", "Jawaban Anda Benar!")
            else:
                messagebox.showerror("Hasil", "Jawaban Anda Salah.")
        except ValueError:
            messagebox.showerror("Error", "Harap masukkan angka yang valid")
        entry_jawaban.delete(0, tk.END)
        generate_problem()

    # Label untuk menampilkan soal
    question = tk.StringVar()
    label_question = tk.Label(root, textvariable=question, font=('Helvetica', 18))
    label_question.pack(pady=10)

    # Input untuk jawaban
    entry_jawaban = tk.Entry(root, font=('Helvetica', 18))
    entry_jawaban.pack(pady=5)

    # Tombol untuk memeriksa jawaban
    tombol_cek = tk.Button(root, text="Cek Jawaban", command=cek_jawaban, font=('Helvetica', 14))
    tombol_cek.pack(pady=10)

    # Dropdown untuk memilih jenis operasi
    tk.Label(root, text="Pilih Operasi:").pack(pady=5)
    operasi_var = tk.StringVar(value="Penjumlahan")
    operasi_dropdown = tk.OptionMenu(root, operasi_var, "Penjumlahan", "Pengurangan", command=generate_problem)
    operasi_dropdown.pack(pady=5)

    # Generate problem awal
    generate_problem()

    root.mainloop()

pembelajaran_matematika()
```

```python
import time

# Fungsi untuk mengupdate waktu mundur
def countdown():
    try:
        total_seconds = int(entry.get())  # Ambil input dari user
    except ValueError:
        messagebox.showerror("Input Error", "Masukkan jumlah detik yang valid.")
        return
    
    if total_seconds <= 0:
        messagebox.showerror("Input Error", "Masukkan waktu yang lebih dari 0 detik.")
        return
    
    while total_seconds > 0:
        mins, secs = divmod(total_seconds, 60)
        time_format = f'{mins:02d}:{secs:02d}'
        label.config(text=time_format)
        root.update()
        time.sleep(1)  # Pause selama 1 detik
        total_seconds -= 1

    label.config(text="Waktu Habis!")
    messagebox.showinfo("Info", "Waktu Habis!")  # Tampilkan pesan waktu habis

# Fungsi untuk memulai hitungan mundur
def start_countdown():
    countdown()

# Inisialisasi tkinter
root = tk.Tk()
root.title("Penghitung Waktu Mundur")

# Input untuk jumlah detik
entry_label = tk.Label(root, text="Masukkan Waktu (dalam detik):")
entry_label.pack()

entry = tk.Entry(root)
entry.pack()

# Tombol untuk memulai hitungan mundur
start_button = tk.Button(root, text="Mulai", command=start_countdown)
start_button.pack()

# Label untuk menampilkan waktu tersisa
label = tk.Label(root, text="00:00", font=("Helvetica", 48))
label.pack()

# Mulai tkinter loop
root.mainloop()
```

```python
from tkinter import messagebox, ttk
import mysql.connector
# Fungsi untuk menghubungkan ke database MySQL
def connect_db():
    return mysql.connector.connect(
        host="localhost",
        user="root",  # Ganti dengan user MySQL Anda
        password="",  # Ganti dengan password MySQL Anda
        database="Pengelolaan_Buku"
    )

# Fungsi untuk menambahkan buku ke database
def add_book():
    title = title_entry.get()
    author = author_entry.get()
    year = year_entry.get()
    
    if title == "" or author == "" or year == "":
        messagebox.showerror("Input Error", "Semua kolom harus diisi.")
        return
    
    try:
        conn = connect_db()
        cursor = conn.cursor()
        cursor.execute("INSERT INTO Buku (Judul_Buku, Pengarang, tahun_terbit) VALUES (%s, %s, %s)", (title, author, year))
        conn.commit()
        conn.close()
        messagebox.showinfo("Success", "Buku berhasil ditambahkan.")
        clear_inputs()
        display_books()
    except Exception as e:
        messagebox.showerror("Database Error", str(e))

# Fungsi untuk menampilkan daftar buku dari database
def display_books():
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM Buku")
    rows = cursor.fetchall()
    conn.close()
    
    for i in book_tree.get_children():
        book_tree.delete(i)
    
    for row in rows:
        book_tree.insert("", tk.END, values=row)

# Fungsi untuk menghapus buku dari database
def delete_book():
    selected_item = book_tree.selection()
    if not selected_item:
        messagebox.showerror("Selection Error", "Pilih buku untuk dihapus.")
        return

    book_id = book_tree.item(selected_item, "values")[0]
    conn = connect_db()
    cursor = conn.cursor()
    cursor.execute("DELETE FROM Buku WHERE Id_buku=%s", (book_id,))
    conn.commit()
    conn.close()
    messagebox.showinfo("Success", "Buku berhasil dihapus.")
    display_books()

# Fungsi untuk memperbarui data buku
def update_book():
    selected_item = book_tree.selection()
    if not selected_item:
        messagebox.showerror("Selection Error", "Pilih buku untuk diperbarui.")
        return

    book_id = book_tree.item(selected_item, "values")[0]
    title = title_entry.get()
    author = author_entry.get()
    year = year_entry.get()

    if title == "" or author == "" or year == "":
        messagebox.showerror("Input Error", "Semua kolom harus diisi.")
        return

    try:
        conn = connect_db()
        cursor = conn.cursor()
        cursor.execute("UPDATE Buku SET Judul_Buku=%s, Pengarang=%s, tahun_terbit=%s WHERE Id_buku=%s", (title, author, year, book_id))
        conn.commit()
        conn.close()
        messagebox.showinfo("Success", "Buku berhasil diperbarui.")
        clear_inputs()
        display_books()
    except Exception as e:
        messagebox.showerror("Database Error", str(e))

# Fungsi untuk membersihkan input
def clear_inputs():
    title_entry.delete(0, tk.END)
    author_entry.delete(0, tk.END)
    year_entry.delete(0, tk.END)

# Membuat antarmuka pengguna dengan Tkinter
root = tk.Tk()
root.title("Aplikasi Pengelolaan Buku")

# Frame untuk input buku
input_frame = tk.Frame(root)
input_frame.pack(pady=20)

tk.Label(input_frame, text="Judul Buku").grid(row=0, column=0, padx=10, pady=10)
tk.Label(input_frame, text="Pengarang").grid(row=1, column=0, padx=10, pady=10)
tk.Label(input_frame, text="Tahun Terbit").grid(row=2, column=0, padx=10, pady=10)

title_entry = tk.Entry(input_frame)
author_entry = tk.Entry(input_frame)
year_entry = tk.Entry(input_frame)

title_entry.grid(row=0, column=1, padx=10, pady=10)
author_entry.grid(row=1, column=1, padx=10, pady=10)
year_entry.grid(row=2, column=1, padx=10, pady=10)

# Tombol untuk fungsi CRUD
button_frame = tk.Frame(root)
button_frame.pack(pady=10)

add_button = tk.Button(button_frame, text="Tambah Buku", command=add_book)
add_button.grid(row=0, column=0, padx=10)

update_button = tk.Button(button_frame, text="Perbarui Buku", command=update_book)
update_button.grid(row=0, column=1, padx=10)

delete_button = tk.Button(button_frame, text="Hapus Buku", command=delete_book)
delete_button.grid(row=0, column=2, padx=10)

# Tabel untuk menampilkan daftar buku
columns = ("Id_buku", "Judul_Buku", "Pengarang", "Tahun_Terbit")
book_tree = ttk.Treeview(root, columns=columns, show="headings", height=8)
book_tree.heading("Id_buku", text="ID")
book_tree.heading("Judul_Buku", text="Judul Buku")
book_tree.heading("Pengarang", text="Pengarang")
book_tree.heading("Tahun_Terbit", text="Tahun Terbit")

book_tree.column("Id_buku", width=50)
book_tree.column("Judul_Buku", width=150)
book_tree.column("Pengarang", width=100)
book_tree.column("Tahun_Terbit", width=80)

book_tree.pack(pady=10)

# Menghubungkan ke database dan menampilkan daftar buku
display_books()

# Jalankan aplikasi Tkinter
root.mainloop()
```
