```python
# Import Library
import matplotlib.pyplot as plt #membuat grafik
import matplotlib.ticker as ticker #membuat grafik
import pandas as pd
```

```python
# Membaca file Excel
df = pd.read_excel('Sample_Superstore.xlsx')

# Menampilkan 5 baris pertama
df.head(10)
```

```python
df = df.drop(columns=['Row ID'])
```

```python
df.head()
```

```python
# Konversi tanggal Order menjadi tanggal format pandas 
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

# Konversi tanggal pengiriman (Ship Date) menjadi format datetime pandas
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')

# NB: Date harus diubah ke formatnya pandas supaya bisa diurutkan
```

```python
df
```

```python
#mengururtkan data tanggal order 
df.sort_values(by = ['Order Date'], inplace = True, ascending = True)
```

```python
#menampilkan hanya menampilkan kolom tertentu
df.filter(['Order ID','Order Date','Ship Date', 'Ship Mode'])
```

```python
#Menamapilkan data costumer dengan nama Darren Powers
df.loc[(df['Customer Name'] == 'Darren Powers')]
```

```python
#Menampilkan data dengan dua syarat yaitu customer dan ship modenya standard class
df.loc[((df['Customer Name'] == 'Darren Powers')&(df['Ship Mode'] == 'Standard Class'))]
```

```python
temantersayang = ["Patrick O'Donnell", 'Darren Powers', 'Sule']
df.loc[(df['Customer Name'].isin(temantersayang))]
```

```python
#menampilkan tabel sales yang kota nya di houston 
#tapi segment nya bukan Consumer ("bukan" ditandai dengan 1=) 
df.loc[((df['City'] == 'Houston') & (df['Segment'] != 'Consumer'))]

#Melihat data unik di kolom Ship Mode 
print(df['Ship Mode'].unique())

#Melihat data unik di kolom Category 
print(df['Category'].unique())

#Melihat data unik di kolom SubCategory 
print(df['Sub-Category'].unique())
```

```python
#Analisis I.
# mekibat jumlab oskee esnivalan 10 State urut dari yang terkecil

stateterkecil = df.filter(['State', 'Sales']) 
stateterkecil.groupby(['State']).sum()['Sales'].nsmallest(10)

#melakukan group of per-state, kemudian salesnya per state di jumlahkan (sum) 
#nsmallest(10) artinya diambil 10 terkecil
```

```python
#Analisis II.
#melihat jumlah rekap penjualan per sub kategori
sales_psktg = df.filter(['Sub-Category', 'Sales'])
sales_psktg.groupby(['Sub-Category']).sum() ['Sales']
```

```python
#Analisis II cara Grafik
#Membuat grafik rekap sales per sub category barang
plt.rcParams['figure.figsize']=(15,3) #mengatur ukuran grafik (cm) 
plt.bar(df['Sub-Category'], df [ 'Sales']); #membuat bar dari df
plt.rcParams.update({'font.size':10}); #mengatur ukuran huruf 
plt.xlabel('Sub_Category'); #memberi Label sumbu X
plt.ylabel('Sales');  #memberi Label sumbu y
```

```python
#menampilkan 20 orang customer yang paling banyak keluarkan uang di perusahaan kita :

#group_by-sum meng-group data per orang sekaligus salesnya dijumlah 
Top_customers=df[['Customer Name', 'Sales']].groupby (["Customer Name"]).sum()

#sort+values: mengurutkan berdasarkan kolom tertentu
#ascending=False urut besar ke kecil
Top_customers = Top_customers.sort_values ('Sales', ascending=False)

#head(20): mengambil 20 data pertama dari atas
Top_customers = Top_customers.head(20)

#formating nilai sales Lebih rapi dgn hanya 2 digit di belakang koma 
Top_customers = Top_customers [["Sales"]].round(2)

#memberi index pada Top_customers
Top_customers.reset_index(inplace=True)
Top_customers
```

```python
#hanya bisa dijalankan setelah Top_customer diproses
plt.figure(figsize = (15,5)) #panjang=15 dan tinggi grafik=5

#Menggambar grafik dengan warna green (hijau)
#menuliskan judul grafik dengan ukuran font 18
plt.title("Most Valuable Customers (2015-2019)", fontsize=18)
plt.bar(Top_customers["Customer Name"], 
        Top_customers ["Sales"],color= '#99ff99', edgecolor='green', 
        linewidth = 1)
plt.xlabel("Customers", fontsize=15) # Label sumbu X
plt.ylabel("Revenue", fontsize=15) # Label sumbu y
plt.xticks(fontsize=12, rotation=90) #Label sumbu x dirotasi tegak
plt.yticks (fontsize=12)

#Mencetak setiap angka sales dalam batang grafik:
for k,v in Top_customers ["Sales"].items():
    plt.text(k,v-8000, '$'+ str(v), fontsize=12, rotation=90,color='k', 
             horizontalalignment='center');
```

```python
#BUATLAH:
#1. Tabel & Grafik yang menampilkan 10 Kota dengan Sales Terbanyak
# Mengelompokkan data berdasarkan kota dan menghitung total penjualan
city_sales = df.groupby('City')['Sales'].sum()
# Mengambil 10 kota dengan penjualan terbanyak
top_10_cities = city_sales.nlargest(10)
print(top_10_cities)

# Membuat grafik batang
plt.figure(figsize=(10, 6))
top_10_cities.plot(kind='bar', color='skyblue')

# Menambahkan label dan judul
plt.xlabel('City')
plt.ylabel('Sales')
plt.title('10 Kota dengan Sales Terbanyak')
plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x untuk keterbacaan

# Menampilkan grafik
plt.tight_layout()  # Menyesuaikan layout
plt.show()


```

```python
#2. Tabel & Grafik yang menampilkan 10 State dengan Sales Terbanyak
# Mengelompokkan data berdasarkan State dan menghitung total penjualan
state_sales = df.groupby('State')['Sales'].sum()
# Mengambil 10 kota dengan penjualan terbanyak
top_10_state = state_sales.nlargest(10)
print(top_10_state)

# Membuat grafik batang
plt.figure(figsize=(10, 6))
top_10_state.plot(kind='bar', color='Green')

# Menambahkan label dan judul
plt.xlabel('State')
plt.ylabel('Sales')
plt.title('10 State dengan Sales Terbanyak')
plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x untuk keterbacaan

# Menampilkan grafik
plt.tight_layout()  # Menyesuaikan layout
plt.show()
```

```python
#3. Tabel & Grafik yang menampilkan Sales 10 Sub Category Barang Terbanyak
# Mengelompokkan data berdasarkan sub-kategori dan menghitung total penjualan
sub_category_sales = df.groupby('Sub-Category')['Sales'].sum()

# Mengambil 10 sub-kategori dengan penjualan terbanyak
top_10_sub_categories = sub_category_sales.nlargest(10)

# Menampilkan tabel
print(top_10_sub_categories)

# Membuat grafik batang
plt.figure(figsize=(12, 8))
top_10_sub_categories.plot(kind='bar', color='Red')

# Menambahkan label dan judul
plt.xlabel('Sub-Category')
plt.ylabel('Sales')
plt.title('10 Sub-Kategori Barang dengan Penjualan Terbanyak')
plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x untuk keterbacaan

# Menampilkan grafik
plt.tight_layout()  # Menyesuaikan layout
plt.show()
```

```python
#4. Tabel & Grafik yang menampilkan 10 Customer dengan profit terbaik
# Mengelompokkan data berdasarkan pelanggan dan menghitung total profit
customer_profit = df.groupby('Customer Name')['Profit'].sum()
# Mengambil 10 pelanggan dengan profit tertinggi
top_10_customers = customer_profit.nlargest(10)
# Menampilkan tabel
print(top_10_customers)

# Membuat grafik batang
plt.figure(figsize=(12, 8))
top_10_customers.plot(kind='bar', color='Blue')

# Menambahkan label dan judul
plt.xlabel('Customer')
plt.ylabel('Profit')
plt.title('10 Customer dengan Profit Terbaik')
plt.xticks(rotation=45, ha='right')  # Memutar label sumbu x untuk keterbacaan

# Menampilkan grafik
plt.tight_layout()  # Menyesuaikan layout
plt.show()
```

```python
import seaborn as sns

#Mengcopy tabeldf ke newdf, sehingga step Lanjutannya tidak mengganggu/merusak df 
newdf=df.copy()

#Step 2: Aggregate sales by month to see the monthly sales trend 
monthly_sales = newdf ['Sales'].resample('ME').sum()

#Step 1: Plot the monthly sales trend using matplotlib and seaborn
plt.figure(figsize=(12, 6)) 
sns.lineplot(x=monthly_sales.index, y=monthly_sales.values) 
plt.title('Monthly Sales Trend - Sample Superstore')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()

# Step 4: Optional Further analysis (yearly, quarterly trends)
yearly_sales = df.resample('Y').sum()['Sales']
plt.figure(figsize=(10, 5))
sns.barplot(x=yearly_sales.index.year, y=yearly_sales.values, palette='viridis')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```
