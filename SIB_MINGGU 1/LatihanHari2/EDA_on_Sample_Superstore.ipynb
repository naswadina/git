```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('SampleSuperstore.csv')
```

```python
data
```

```python
data.describe(include='all')  # ringkasan statistik dari data
```

```python
data.hist(bins=10, figsize=(18,10)) #membuat histogram dari semua kolom numerik
```

```python
numeric_data = data.select_dtypes(include=[np.number]) #Memilih hanya kolom-kolom numerik dari data
plt.figure(figsize=(4,4)) #Mengatur ukuran gambar/plot sebesar 4x4
sns.heatmap(numeric_data.corr(), annot=True) #Membuat heatmap untuk menampilkan korelasi
```

## [1] Analysis based on Ship Mode

```python
# Membuat plot pasangan (pairplot) untuk semua kombinasi kolom numerik dalam data
sns.pairplot(data, hue='Ship Mode')
```

```python
#Membuat count plot (bar plot) yang menunjukkan jumlah pengamatan (frekuensi) untuk setiap kategori dalam kolom Ship Mode
sns.countplot(x='Ship Mode', data=data, palette='tab10') 
```

```python
#Mengelompokkan data berdasarkan kolom Ship Mode
#Menghitung rata-rata dari kolom-kolom numerik Sales, Profit, dan Discount
data_shipmode = data.groupby(['Ship Mode'])[['Sales', 'Profit', 'Discount']].mean()
data_shipmode
```

```python
colors_4 = ['magenta','yellow','orange','red']
colors_3 = ['green', 'blue','cyan']
```

```python
#Membuat plot pie (diagram lingkaran) untuk setiap kolom dalam data_shipmode yaitu Sales, Profit, dan Discount
data_shipmode.plot.pie(subplots=True, figsize=(20,10), labels=data_shipmode.index, autopct='%1.1f%%', explode=(0.01,0.01,0.01,0.01), colors=colors_4)
plt.show()
```

```python
#Melihat profit dan penjualan bervariasi untuk setiap kategori pengiriman.
#Mengidentifikasi kategori dengan profit tertinggi dan terendah, serta penjualan terkait dengan profit tersebut
data_shipmode.sort_values('Profit')[['Profit','Sales']].plot(kind='bar', figsize=(10,7))
```

```python
data_shipmode.sum(axis=0) #Menghitung jumlah total dari setiap kolom dalam data
```

```python
#menormalisasi atau menskalakan nilai Sales dan Profit
#Membuat kolom baru Sales% dalam dataframe data_shipmode.
#Mengisi kolom ini dengan nilai Sales dibagi dengan 9.285
data_shipmode['Sales%'] = data_shipmode['Sales']/9.285

#Membuat kolom baru Profit% dalam dataframe data_shipmode.
#Mengisi kolom ini dengan nilai Profit dibagi dengan 1.181
data_shipmode['Profit%'] = data_shipmode['Profit']/1.181
```

```python
#Mengurutkan data data_shipmode berdasarkan kolom Profit% dalam urutan menaik.
data_shipmode.sort_values('Profit%')
```

## Sales
### Lowest sales: Standard Class(24.51%)
### Highest sales: Same Day(25.46%)

## Profit
### Lowest profit: Standard Class(23.28%)
### Highest profit: First class(26.96%)

#### Though there is very little differences between the classes, still, maximum and minimum can be ascertained on seeing these analysis.
