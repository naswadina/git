```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

#Data penjualan minggu 1 sampai 5
x=[[1],[2],[3],[4],[5]]

#x adalah minggu, sifat : independent
y=[[150], [157], [162], [166], [177]]

#y adalah data penjualan di minggu tersebut. sifat : dependent
x_train=x
y_train=y
lm = LinearRegression()
lm.fit(x_train,y_train) #fase training

import pandas as pd
x_test = [[6]] #data yang akan diprediksi, penjualan ke 6
p = lm.predict(x_test)

#fase prediksi
print('Angka ramalan untuk penjualan minggu ke 6 : ',p)

#hasil prediksi
#prepare plot
pb = lm.predict(x_train)
plt.scatter(x_train,y_train)
plt.plot(x_train,pb,color='red',linewidth=2)
plt.xlabel('Minggu')
plt.ylabel('Penjualan')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_percentage_error

# Data penjualan bulan 1 sampai 5
x = [[1], [2], [3], [4], [5]]  # Bulan (variabel independen)
y = [[220], [250], [210], [300], [325]]  # Penjualan (variabel dependen)

# Membuat model regresi linier
model = LinearRegression()
model.fit(x, y)  # Training model

# Prediksi untuk bulan yang belum diketahui (bulan 2 sampai 5)
x_test = [[2], [3], [4], [5]]  # Bulan ke-2, 3, 4, 5
y_pred = model.predict(x_test)

# Menampilkan hasil prediksi
for i, pred in enumerate(y_pred, start=2):
    print(f'Prediksi penjualan bulan ke-{i}: {pred[0]}')

# Menghitung MAPE (Mean Absolute Percentage Error) hanya untuk bulan ke-2 hingga ke-5
y_actual = np.array([250, 210, 300, 325]).reshape(-1, 1)  # Nilai aktual bulan 2-5
mape = mean_absolute_percentage_error(y_actual, y_pred) * 100
print(f'Nilai MAPE: {mape:.2f}%')

# Visualisasi hasil regresi
x_train = x
y_train = y
plt.scatter(x_train, y_train, color='blue', label='Data aktual')
plt.plot(x_train, model.predict(x_train), color='red', linewidth=2, label='Garis regresi')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.title('Regresi Linier Penjualan')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error

# Data penjualan bulan 1 sampai 5
x = np.array([[1], [2], [3], [4], [5]])  # Bulan (variabel independen)
y = np.array([220, 250, 210, 300, 325])  # Penjualan (variabel dependen)

# Membuat model Random Forest Regressor
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x, y)  # Training model

# Prediksi untuk bulan yang belum diketahui (bulan 2 sampai 5)
x_test = np.array([[2], [3], [4], [5]])  # Bulan ke-2, 3, 4, 5
y_pred = model.predict(x_test)

# Menampilkan hasil prediksi
for i, pred in enumerate(y_pred, start=2):
    print(f'Prediksi penjualan bulan ke-{i}: {pred:.2f}')  # Hanya mencetak pred

# Menghitung MAPE (Mean Absolute Percentage Error) hanya untuk bulan ke-2 hingga ke-5
y_actual = np.array([250, 210, 300, 325])  # Nilai aktual bulan 2-5
mape = mean_absolute_percentage_error(y_actual, y_pred) * 100
print(f'Nilai MAPE: {mape:.2f}%')

# Visualisasi hasil regresi
plt.scatter(x, y, color='blue', label='Data aktual')
plt.plot(x, model.predict(x), color='green', linewidth=2, label='Garis regresi')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.title('Random Forest Regressor - Penjualan')
plt.legend()
plt.show()
```
